```python
import pandas as pd
import geopandas as gpd
from sqlalchemy import create_engine
import os
```

```python
user = os.environ.get("DISTR_DB_USER")
password = os.environ.get("DISTR_DB_PASS")
host = os.environ.get("DISTR_DB_HOST")
database = os.environ.get("DISTR_DB_NAME")
```

```python
db_connection_url = f'postgresql://{user}:{password}@{host}:5432/{database}'
con = create_engine(db_connection_url)
```

```python
# read data
pdb2023tr = pd.read_csv('../data/pdb2023tr.csv')
```

```python
pdb2023tr = pdb2023tr[~pdb2023tr['State'].isin((72,))]
```

```python
pdb2023tr.to_sql('pdb2023tr', con = con, chunksize = 100, method = 'multi', index = False)
```

```python
tracts_geom = gpd.read_file('../data/tl2023_us_tract_2023/US_tract_2023.shp')
```

```python
tracts_geom.to_postgis('tracts_2023', con = con, index = False)
```
