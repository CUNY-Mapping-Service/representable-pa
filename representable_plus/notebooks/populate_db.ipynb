```python
import pandas as pd
import geopandas as gpd
from sqlalchemy import create_engine
import os
```

```python
user = os.environ.get("DISTR_DB_USER")
password = os.environ.get("DISTR_DB_PASS")
host = os.environ.get("DISTR_DB_HOST")
database = os.environ.get("DISTR_DB_NAME")
```

```python
db_connection_url = f'postgresql://{user}:{password}@{host}:5432/{database}'
con = create_engine(db_connection_url)
```

```python
# read data
# https://www2.census.gov/adrm/PDB/2023/pdb2023tr.csv
pdb2023tr = pd.read_csv('../data/pdb2023tr.csv')
```

```python
pdb2023tr = pdb2023tr[pdb2023tr['State'].isin((42,))]
len(pdb2023tr)
```

```python
pdb2023tr.to_sql('pdb2023tr', con = con, chunksize = 1000, index = False, if_exists='replace')
```

```python
# https://www.census.gov/cgi-bin/geo/shapefiles/index.php?year=2025&layergroup=Census+Tracts
tracts_geom = gpd.read_file('../data/tl_2025_42_tract/tl_2025_42_tract.shp')
```

```python
tracts_geom.to_postgis('tracts_2023', con = con, index = False)
```
